```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
# Loading Data
df0 = pd.read_csv("train_v9rqX0R.csv")
```

```python
# Checking the dataframe
df0.head()
```

```python
# Keeping a copy
df1 = df0.copy()

# Dimension of df1
print(df1.shape)

# Basic info for the dataset
df1.info()
```

```python
# Getting inferential Statistics
df1.describe()
```

```python
# Check for missing values in dataframe
print('Total count of missing values:', df1.isna().sum().sum())

# Display missing values per column in dataframe
print('Missing values per column:')
df1.isna().sum()
```

```python
df1[df1["Outlet_Size"].isnull()]["Outlet_Identifier"].unique()
```

```python
# Drop rows where "Outlet_Size" column contains null values
df1.dropna(subset=["Outlet_Size"], inplace=True)
df1.info()
```

```python
df1.isna().sum()
```

```python
# Calculate mean Item_Weight for each item type
item_type_mean = df1.groupby("Item_Type")["Item_Weight"].transform("mean")
item_type_mean
```

```python
# Calculate median Item_Weight for each item type
item_type_median = df1.groupby("Item_Type")["Item_Weight"].median()
item_type_median
```

```python
# Fill NaN values in Item_Weight with mean for each item type
df1["Item_Weight"] = df1["Item_Weight"].fillna(item_type_mean)
```

```python
df1.info()
```

```python
# Using pairplot to check if there exists a linear relationship
sns.pairplot(df1)
```

```python
# Create figure and axes objects
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2 row, 2 columns for 4 plots

# checking relationship between Outlet_Type and Item_Outlet_Sales
sales_by_outlet_type = df1.groupby("Outlet_Type").sum()["Item_Outlet_Sales"]

sns.barplot(
    data=sales_by_outlet_type.reset_index(),
    x="Outlet_Type",
    y="Item_Outlet_Sales",
    ax=axes[0,0]
)

axes[0,0].tick_params(axis='x', rotation=45)

# checking relationship between Outlet_Location_Type and Item_Outlet_Sales
sales_by_outlet_location_type = df1.groupby("Outlet_Location_Type").sum()["Item_Outlet_Sales"]

sns.barplot(
    data=sales_by_outlet_location_type.reset_index(),
    x="Outlet_Location_Type",
    y="Item_Outlet_Sales",
    ax=axes[0,1]
)

# checking relationship between Outlet_Size and Item_Outlet_Sales
sales_by_outlet_size = df1.groupby("Outlet_Size").sum()["Item_Outlet_Sales"]

sns.barplot(
    data=sales_by_outlet_size.reset_index(),
    x="Outlet_Size",
    y="Item_Outlet_Sales",
    ax=axes[1,0]
)

# checking relationship between Item_Fat_Content and Item_Outlet_Sales
sales_by_item_fat_content = df1.groupby("Item_Fat_Content").sum()["Item_Outlet_Sales"]

sns.barplot(
    data=sales_by_item_fat_content.reset_index(),
    x="Item_Fat_Content",
    y="Item_Outlet_Sales",
    ax=axes[1,1]
)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
